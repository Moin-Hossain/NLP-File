```python
#Imports and defaults
import json
import ollama
import networkx as nx
from tqdm.notebook import tqdm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict
from datasets import load_dataset
import matplotlib.colors as mcolors
from matplotlib.patches import Patch
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')
from bokeh.plotting import show

defaults = dict(width=1000, height=800, padding=0.1,
                xaxis=None, yaxis=None)
hv.opts.defaults(
    opts.EdgePaths(**defaults), opts.Graph(**defaults), opts.Nodes(**defaults))
```

```python
#Here we construct the biparite graph by loading our edgelist constructed in the previous notebook via LLM. We do so focusing actually on the "from" which is the company, and the "industry_type", so the nodes are companies and industries. We also implement weight,
#which is determined by the occurence of technologies a company has attatched to a certain industry which is also determined by the LLM.
with open("/content/edges_list.json", "r", encoding='utf-8') as f:
    edges = json.load(f)

B = nx.Graph()
companies = set()
industry_types = set()
edge_weights = defaultdict(int)

print("Building bipartite graph with weights...")
for edge in tqdm(edges, desc="Processing Edges"):
    company = edge.get('from')
    industry_type = edge.get('industry_type')
    weight = edge.get('weight', 1)
    if not company or not industry_type:
        continue
    companies.add(company)
    industry_types.add(industry_type)
    B.add_edge(company, industry_type, weight=B[company][industry_type]['weight'] + weight if B.has_edge(company, industry_type) else weight)

print(f"\nNumber of companies: {len(companies)}")
print(f"Number of industry types: {len(industry_types)}")
print(f"Number of edges in the bipartite graph: {B.number_of_edges()}")

#Here we project onto the company layer using weighted_projected_graph
print("\nProjecting bipartite graph onto company layer...")
company_graph = nx.bipartite.weighted_projected_graph(B, companies)
print(f"Number of nodes in company graph: {company_graph.number_of_nodes()}")
print(f"Number of edges in company graph: {company_graph.number_of_edges()}")



#We can see that we get 59471 edges between 997 nodes. This can be interpreted that there are 59471 connections between the 997 companies via shared industries. This means that we have a highly connected network, meaning that many of the companies are expected to be operating across the same industries.
```

```python
#We initialize a trimming process by defining that we want to remove edges that have edge weight below the 20% threshold, this means in practice that the 20% least weighted edges will be removed from the graph as long as they are below the threshold we establish.
def trim_edges(graph, percentile=20):
    if not graph.edges(data=True):
        print("The graph has no edges to trim.")
        return graph

    weights = [data['weight'] for _, _, data in graph.edges(data=True)]
    threshold = np.percentile(weights, percentile)
    print(f"Trimming edges with weights below the {percentile}th percentile (threshold: {threshold:.4f})")

    trimmed_graph = nx.Graph()
    trimmed_graph.add_nodes_from(graph.nodes(data=True))
    for u, v, data in graph.edges(data=True):
        if data['weight'] >= threshold:
            trimmed_graph.add_edge(u, v, **data)

    return trimmed_graph
```

```python
# We trim edges with lowest weights with 20% as the goal.
percentile = 20
print("\nTrimming low-weight edges...")
trimmed_company_graph = trim_edges(company_graph, percentile=percentile)
print(f"Number of nodes after trimming: {trimmed_company_graph.number_of_nodes()}")
print(f"Number of edges after trimming: {trimmed_company_graph.number_of_edges()}")

#We calculate centralities on the trimmed graph
print("\nCalculating centrality measures for companies and sectors...")
degree_centrality = nx.degree_centrality(trimmed_company_graph)
betweenness_centrality = nx.betweenness_centrality(trimmed_company_graph)
eigenvector_centrality = nx.eigenvector_centrality(trimmed_company_graph, max_iter=1000)

#We combine centralities
combined_centrality = {
    node: (degree_centrality.get(node, 0) +
           betweenness_centrality.get(node, 0) +
           eigenvector_centrality.get(node, 0)) / 3
    for node in trimmed_company_graph.nodes()
}

#We sort nodes by combined centrality to identify key companies and sectors
sorted_nodes = sorted(combined_centrality.items(), key=lambda x: x[1], reverse=True)

#As we want to depict the relation between sectors and companies and especially look at population we determine that we want a lot of nodes in our plot, hence we create a subgraph focused on this by maintaining the majority.
N = 600
top_nodes = [node for node, _ in sorted_nodes[:N]]

#We create a subgraph with only the top nodes
subgraph = trimmed_company_graph.subgraph(top_nodes).copy()
print(f"\nNumber of nodes in subgraph: {subgraph.number_of_nodes()}")
print(f"Number of edges in subgraph: {subgraph.number_of_edges()}")


#We interpret this as follows, we initially actually didnt trim any values meaning that they all were in our threshold value. However, when we choose to subgraph for 600 nodes we can see that we maintain 51085 edges, so we have left out some of the least connected nodes in the end.
```

```python
#Here we construct the graph. We focus on showing the weighted relation between nodes via transparancy of edges, and also colors to visualize each industries nodes.

#We ensure that companies are related to an industry for for implementing coloring bases on that.
company_to_industry = {node: list(B.neighbors(node)) for node in companies if B.has_node(node)}
for company, industries in company_to_industry.items():
    if company in company_graph.nodes:
        company_graph.nodes[company]['industry_type'] = industries[0]
for company in subgraph.nodes:
    if 'industry_type' not in subgraph.nodes[company] and company in company_to_industry:
        subgraph.nodes[company]['industry_type'] = company_to_industry[company][0]

#We establish colors based on industry type for better viusalization
color_palette = plt.get_cmap('tab20')
unique_industries = list(set(nx.get_node_attributes(subgraph, 'industry_type').values()))
color_map = {industry: color_palette(i / len(unique_industries)) for i, industry in enumerate(unique_industries)}
node_colors = [color_map.get(subgraph.nodes[node].get('industry_type', 'Other'), 'gray') for node in subgraph.nodes()]
legend_elements = [Patch(facecolor=color, label=industry) for industry, color in color_map.items()]

#We repare for visualization
pos = nx.spring_layout(subgraph, k=0.3, iterations=50)
sizes = [combined_centrality[node] * 1000 for node in subgraph.nodes()]
labels = {node: node for node in subgraph.nodes()}

#We retrieve all edge weights for normalization
weights = [data['weight'] for _, _, data in subgraph.edges(data=True)]
min_weight, max_weight = min(weights), max(weights)
edge_alphas = [(data['weight'] - min_weight) / (max_weight - min_weight) for _, _, data in subgraph.edges(data=True)]

#We visualize the network
plt.figure(figsize=(15, 20))
nx.draw_networkx_nodes(subgraph, pos, node_size=sizes, alpha=0.8, node_color=node_colors)
nx.draw_networkx_edges(
    subgraph, pos,
    width=[data['weight'] / 10 for _, _, data in subgraph.edges(data=True)],
    edge_color='gray',
    alpha=edge_alphas
)
nx.draw_networkx_labels(subgraph, pos, labels, font_size=5)

#We add legend to the plot
plt.legend(handles=legend_elements, loc="upper left", title="Industry Types", bbox_to_anchor=(1, 1))

plt.title("Company Relationships based on Shared Industry Types", fontsize=5)
plt.axis('off')
plt.tight_layout()
plt.savefig('company_network.png', dpi=300, bbox_inches='tight')
plt.show()



#Interpreting this visualization shows us that the nodes have an almost cluster-like shape, of course with some exceptions, based on the industry. This means that industries are much seperated, and besides a few nodes that a depicted central in the graph, many nodes have connections primarily inside the
#the business industry. We can also see that central is the software development category. This seems to be the most central business sector, which makes sense when you think about it. While the software development is an industry, one could have perhaps left out this label and seen what would have happened,
#as AI implementation is often partly software development of course, which could have potentially confused the LLM. By looking at the edges we can see that in the most central cluster many of the nodes share a stronger connection than outside of it. This means that companies are more highly connected
#via their industry in this place. Also on node-size that is regarding centrality we can see that the nodes in the middle again are much more central, and that nodes outside are barely of any centrality.
#To sort of encapsulate the essense of this graph in terms of the findings we set out to uncover, one could say that only a few industries actually have a substantial amount of nodes. Many seemingly have very small if any community of companies. This means that to sort of market analyze
#it means either that there is room for more AI-related businesses in those sectors, or for bigger companies to branch out. Or that those industries' main actors should improve their fundamentals to open up for more business related to AI to seek into this, as this would of course be benifical
#on the long run.
```

```python
#Next we want to zoom in on the more central specific nodes, i.e. companies. This was not too specific in the previous graph, so here we use holoviews to also add the labels of centrality in our visualization. We focus on the top 50 nodes.

N = 50
top_nodes = [node for node, _ in sorted_nodes[:N]]
subgraph = trimmed_company_graph.subgraph(top_nodes).copy()
nx.set_node_attributes(subgraph, combined_centrality, 'combined_centrality')


#We prepare the data for labels
label_data = {
    'x': [pos[node][0] for node in subgraph.nodes()],
    'y': [pos[node][1] for node in subgraph.nodes()],
    'Label': list(subgraph.nodes())  # Changed key to 'Label'
}

#We create a HoloViews graph
hv_graph = hv.Graph.from_networkx(subgraph, pos).opts(
    node_size=hv.dim('combined_centrality') * 300,
    node_alpha=0.8,
    node_color='combined_centrality',
    edge_alpha=hv.dim('weight') * 0.1,
    cmap='Blues',
    edge_color='gray',
    tools=['hover'],
    title="Company Relationships based on Industry Types",
    width=1000,
    height=800,
    show_frame=False  # Remove the frame
)

#We add labels to nodes
hv_labels = hv.Labels((label_data['x'], label_data['y'], label_data['Label']),
                      label='Label').opts(text_font_size='7pt', text_color='black')

#We combine the graph and labels
final_plot = hv_graph * hv_labels

#We display the plot
show(hv.render(final_plot))


#Here we can see that close togehter and also of most central we find Oracle of combined centrality 0.192 and Google with 0.238. This means that these are highly connected in the network to ohter nodes, probably because they are active in many industries. Lesser are IBM, Mastek, TMA Solutions, who are about
#0.160 to 0.198 in combined centrality. We can see that these are the main sort of nodes, and then in the top 50 we have a lot of lesser important nodes. To further interpret the centrality we move on and look at specific centrality metrics in the next piece of code.
```

```python
#We establish both the centralities but also how much specific nodes are connected to what industry, by finding their distribution.

#We print top N companies and their combined centralities
print(f"\nTop {25} Companies by Centralities:")
for company, combined_centrality_value in sorted_nodes[:25]:
    print(f"{company}:")
    print(f"  Degree Centrality: {degree_centrality.get(company, 0):.4f}")
    print(f"  Betweenness Centrality: {betweenness_centrality.get(company, 0):.4f}")
    print(f"  Eigenvector Centrality: {eigenvector_centrality.get(company, 0):.4f}")
    print(f"  Combined Centrality: {combined_centrality_value:.4f}")

#W calculate weighted distribution
industry_type_distribution = defaultdict(lambda: defaultdict(float))
for company in top_nodes:
    for industry_type in B.neighbors(company):
        weight = B[company][industry_type].get('weight', 1)
        industry_type_distribution[company][industry_type] += weight

#We print as percentage distribution
print(f"\nTechnology Type Distribution for Top {N} Companies:")
for company in top_nodes:
    print(f"\n{company}:")
    total_weight = sum(industry_type_distribution[company].values())
    sorted_tech = sorted(industry_type_distribution[company].items(), key=lambda x: x[1], reverse=True)
    for tech_type, weight in sorted_tech:
        percentage = (weight / total_weight) * 100
        print(f"  {tech_type}: {percentage:.2f}%")

#We save centrality data in a structured format
centrality_data = {
    'Degree Centrality': degree_centrality,
    'Betweenness Centrality': betweenness_centrality,
    'Eigenvector Centrality': eigenvector_centrality,
    'Combined Centrality': combined_centrality
}

with open('company_centralities.json', 'w') as f:
    json.dump(centrality_data, f, indent=2)

print("\nCentrality data saved to company_centralities.json")


#Here we see that google has 0.59 centrality degree, which means that 59 & of nodes were connected via an industry to google. Immediately this makes google look like a very central company in terms of AI markets, but if we look at Betweeness which is 0.06 it becomes apparent that as a "bridge" between nodes
#Google is not that imporant. In other words it doesnt exclusively connect many nodes to other industries. Even worse is the centrality when calculating the eigenvector centrality, which takes into account the importance of the connected nodes. Here we see that it is 0.07 meaning that this node is not
#connected to many other important nodes, hence its own importance is of lesser value. This goes for Google, IBM, and the other top companies. This indicates on a general level that no company is sort of gatekeeper or gateway to other industries. It also indicates that while we had many edges, they are not
#very centralized many of them.
```
